```python
df = df.dropna().reset_index(drop = True).copy()
```

```python
from sklearn.cluster import KMeans
```

```python
# select the columns for clustering
X = df.iloc[:, 2:]
```

```python
# list to store the inertia values
list_inertia = []
```

```python
print(X.dtypes)
```

```python
X = X.apply(pd.to_numeric, errors='ignore')
```

```python
from sklearn.cluster import KMeans

# Iterate over each column in X
for column in X.columns:
    try:
        # Select only the current column for clustering
        column_data = X[[column]]

        # Attempt to apply KMeans clustering
        kmeans = KMeans(n_clusters=2, random_state=0).fit(column_data)

        # If clustering is successful, print the column name
        print("No error for column:", column)

    except Exception as e:
        # If an error occurs, print the column name and the error message
        print("Error for column:", column)
        print("Error message:", str(e))
```

```python
X = X.drop(['ISO2', 'ISO3', 'Indicator', 'Unit', 'Source', 'CTS_Code', 'CTS_Name', 'CTS_Full_Descriptor'], axis=1)
```

```python
# range of cluster numbers
k_range = range(1, 11)

# calculate inertia
for k in k_range:
    kmeans = KMeans(n_clusters = k, n_init = 10, random_state = 84)
    kmeans.fit(X)
    list_inertia.append(kmeans.inertia_)
```

```python
# plotting

import matplotlib.pyplot as plt
plt.figure(figsize = (5, 3), facecolor = "white")
plt.plot(k_range, list_inertia, marker = 'o', linestyle = '-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```

```python
n_clusters = 3

# clustering
kmeans = KMeans(n_clusters = n_clusters, n_init = 10, random_state = 84)
X['Cluster'] = kmeans.fit_predict(X)

# Display the clustered data
display(X)
```
